```python
# import pymatgen
# from pymatgen.core.composition import Composition as pmcomp
# from pymatgen.core.structure import Structure as pmstruc
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Extracting data
doping_levels = ["0", "025", "05", "055", "06", "065", "07", "075", "08", "085", "09", "095", "1"]
doping_levels_num = [0, 0.25, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]

infile = "../../shape-relax-metallic/stats.txt"

df = pd.read_table(infile)
df['doping'] = doping_levels_num
print(df.keys())
df
```

```python
### Plotting magnetism ###
WIDTH = 7
HEIGHT = 3.5
YLIMS = None
XLIMS = [0,1]
X_AXIS = 'Avg Ta $d$-state'
Y_AXIS = 'Mag. Moment ($\mu_b$)'
LABEL_FONT = 17

fig = plt.figure(figsize=(WIDTH,HEIGHT))


tot = df['Mag tot']
plt.plot(df['doping'], tot, color='silver',label="Total")
plt.plot(df['doping'], df['Mag s'], 'ko-',label="s (Mg, Ta)")
plt.plot(df['doping'], df['Mag p'], 'ro-',label="p (O)")
plt.plot(df['doping'], df['Mag d'], 'bo-',label="d (Ta)")

zeros = np.zeros(tot.size)
plt.fill_between(df['doping'], tot, where=tot>=zeros, interpolate=True, color='silver')

# Chance font size
plt.gca().xaxis.label.set_fontsize(LABEL_FONT)
plt.gca().yaxis.label.set_fontsize(LABEL_FONT)
plt.gca().tick_params(labelsize=LABEL_FONT)
# plt.gca().get_yticklabels().fontsize(LABEL_FONT)

plt.grid(True)
plt.xlabel(X_AXIS)
plt.ylabel(Y_AXIS)
plt.xlim(XLIMS)
plt.ylim(YLIMS)
#plt.legend(loc='upper right',ncol=1, fancybox=True, framealpha=0.5) 
plt.legend(fancybox=True, framealpha=0.5, fontsize=LABEL_FONT) 

plt.show()

## Saving plot
fig.savefig('mag_cdoped' + '.png', bbox_inches='tight', dpi=500)
fig.savefig('mag_cdoped-vec' + '.pdf', bbox_inches='tight')
```
